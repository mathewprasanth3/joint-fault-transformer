# Step 1 — Data Exploration
**joint-fault-transformer**

Goal: Load one `.mat` file from the ORION-AE dataset, inspect its structure, visualise all 4 channels, and confirm the label parsing works.

Commit after this step: `explore raw ORION-AE signals`

## 0. Imports

````python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from collections import defaultdict
import re

print('All imports OK')
````

## 1. Point to your data

````python
DATA_DIR = Path('../data/raw/ORION_AE_acoustic_emission_multisensor_datasets_bolts_loosening')

# rglob searches recursively through all campaign subfolders (B/C/D/E/F and torque subfolders)
all_files = sorted(DATA_DIR.rglob('*.mat'))
print(f'Total .mat files found: {len(all_files)}')
print('\nFirst 5 filenames:')
for f in all_files[:5]:
    print(' ', f.name)
````

## 2. Filename parsing — extract label automatically

The filename encodes everything we need:
```
salves_out_05cNm_B_Fs5MHz_Tf1s_2019-04-02-16-22-00_1.mat
              ^    ^
              |    |__ Campaign (B, C, D, E, F)
              |_______ Loosening level (05cNm = Class 1)
```
Class mapping: 05->1, 10->2, 20->3, 30->4, 40->5, 50->6, 60->7

````python
TORQUE_TO_CLASS = {
    '05': 1,
    '10': 2,
    '20': 3,
    '30': 4,
    '40': 5,
    '50': 6,
    '60': 7,
}

TORQUE_TO_NM = {
    '05': '0.05 Nm (nearly free)',
    '10': '0.10 Nm (very loose)',
    '20': '0.20 Nm (significantly loose)',
    '30': '0.30 Nm (half nominal)',
    '40': '0.40 Nm (noticeable preload loss)',
    '50': '0.50 Nm (slightly reduced preload)',
    '60': '0.60 Nm (fully tight)',
}

def parse_filename(filepath):
    name = Path(filepath).name
    match = re.search(r'_(\d{2})cNm_([BCDEF])_', name)
    if not match:
        raise ValueError(f'Could not parse filename: {name}')
    torque_str = match.group(1)
    campaign   = match.group(2)
    return {
        'filename'   : name,
        'torque_str' : torque_str,
        'class_label': TORQUE_TO_CLASS[torque_str],
        'campaign'   : campaign,
        'torque_nm'  : TORQUE_TO_NM[torque_str],
    }

# Test on first 5 files
print(f'{"Filename":<60} {"Campaign":<10} {"Class":<7} {"Physical State"}')
print('-' * 110)
for f in all_files[:5]:
    info = parse_filename(f)
    print(f'{info["filename"]:<60} {info["campaign"]:<10} {info["class_label"]:<7} {info["torque_nm"]}')
````

## 3. Dataset overview — files per campaign and level

````python
counts    = defaultdict(lambda: defaultdict(int))
for f in all_files:
    info = parse_filename(f)
    counts[info['campaign']][info['torque_str']] += 1

torques   = ['05', '10', '20', '30', '40', '50', '60']
campaigns = ['B', 'C', 'D', 'E', 'F']

print(f'{"Campaign":<12}', end='')
for t in torques:
    print(f'{t}cNm'.center(8), end='')
print(f'{"TOTAL":>8}')
print('-' * 70)

for c in campaigns:
    print(f'{c:<12}', end='')
    total = 0
    for t in torques:
        n = counts[c][t]
        total += n
        marker = str(n) if n > 0 else '---'
        print(marker.center(8), end='')
    print(f'{total:>8}')

print('-' * 70)
print(f'{"TOTAL":<12}', end='')
grand = 0
for t in torques:
    n = sum(counts[c][t] for c in campaigns)
    grand += n
    print(str(n).center(8), end='')
print(f'{grand:>8}')
print(f'\nNote: Campaign C is missing 40cNm level (known dataset gap)')
````

## 4. Load one .mat file and inspect its structure

````python
sample_file = next(DATA_DIR.rglob('*05cNm_B*.mat'))
info = parse_filename(sample_file)

print(f'Loading  : {sample_file.name}')
print(f'Label    : Class {info["class_label"]} -- {info["torque_nm"]}')
print(f'Campaign : {info["campaign"]}')

mat = scipy.io.loadmat(sample_file)

print(f'\n--- .mat file keys ---')
for key, val in mat.items():
    if not key.startswith('_'):
        if hasattr(val, 'shape'):
            print(f'  {key}: shape={val.shape}, dtype={val.dtype}')
        else:
            print(f'  {key}: {type(val).__name__} = {val}')
````

## 5. Build the 4-channel signal matrix

````python
# Each channel stored as separate key A/B/C/D with shape (N, 1)
signals = np.stack([
    mat['A'].squeeze(),
    mat['B'].squeeze(),
    mat['C'].squeeze(),
    mat['D'].squeeze(),
], axis=0).astype(np.float32)   # final shape: (4, N)

N_SAMPLES  = signals.shape[1]
FS         = 5_000_000
DURATION_S = N_SAMPLES / FS

print(f'Signal matrix shape : {signals.shape}  -> (4 channels x {N_SAMPLES:,} samples)')
print(f'Sampling frequency  : {FS/1e6:.0f} MHz')
print(f'Duration            : {DURATION_S:.3f} seconds')
print(f'Time resolution     : {1/FS*1e6:.1f} us per sample')
print()

CHANNEL_NAMES  = ['A -- Micro80 (AE, 100-300 kHz)',
                  'B -- F50A    (AE, 50-400 kHz)',
                  'C -- Micro200HF (AE, 200 kHz+)',
                  'D -- Velocimeter (laser vibrometer)']
CHANNEL_COLORS = ['#2196F3', '#4CAF50', '#FF9800', '#9C27B0']
CHANNEL_LABELS = ['Ch A  Micro80\n(AE 100-300 kHz)',
                  'Ch B  F50A\n(AE 50-400 kHz)',
                  'Ch C  Micro200HF\n(AE 200 kHz+)',
                  'Ch D  Velocimeter\n(validation only)']
MODEL_INPUT    = [True, True, True, False]

for i, name in enumerate(CHANNEL_NAMES):
    ch = signals[i]
    print(f'Channel {name}')
    print(f'  min={ch.min():.4f} mV  max={ch.max():.4f} mV  '
          f'mean={ch.mean():.6f} mV  std={ch.std():.4f} mV')
````

## 6. Visualise all 4 channels — full 1 second

````python
time_s = np.arange(N_SAMPLES) / FS

fig, axes = plt.subplots(4, 1, figsize=(16, 10), sharex=True)
fig.suptitle(
    f'ORION-AE -- All 4 Channels -- Class {info["class_label"]} '
    f'({info["torque_str"]} cNm, {info["torque_nm"]}) -- Campaign {info["campaign"]}',
    fontsize=13, fontweight='bold'
)

for i, ax in enumerate(axes):
    ax.plot(time_s, signals[i], color=CHANNEL_COLORS[i], linewidth=0.3, alpha=0.85)
    ax.set_ylabel(CHANNEL_LABELS[i], fontsize=9)
    ax.set_ylim(signals[i].min() * 1.1, signals[i].max() * 1.1)
    tag = 'MODEL INPUT' if MODEL_INPUT[i] else 'VALIDATION ONLY'
    ax.annotate(tag, xy=(0.01, 0.88), xycoords='axes fraction',
                fontsize=8, color='white',
                bbox=dict(boxstyle='round,pad=0.3',
                          fc='#1565C0' if MODEL_INPUT[i] else '#6A1B9A',
                          alpha=0.85))
    ax.grid(True, alpha=0.2)

axes[-1].set_xlabel('Time (seconds)', fontsize=10)
plt.tight_layout()
plt.savefig('../outputs/figures/01_all_channels_overview.png', dpi=150, bbox_inches='tight')
plt.show()
print('Saved -> outputs/figures/01_all_channels_overview.png')
````

## 7. Zoom into 10ms — see individual AE burst events

````python
ZOOM_MS   = 10
zoom_samp = int(ZOOM_MS / 1000 * FS)
time_ms   = np.arange(zoom_samp) / FS * 1000

fig, axes = plt.subplots(4, 1, figsize=(16, 10), sharex=True)
fig.suptitle(
    f'ORION-AE -- First {ZOOM_MS}ms -- Class {info["class_label"]} '
    f'({info["torque_str"]} cNm) -- AE bursts visible',
    fontsize=13, fontweight='bold'
)

for i, ax in enumerate(axes):
    ax.plot(time_ms, signals[i, :zoom_samp],
            color=CHANNEL_COLORS[i], linewidth=0.6)
    ax.set_ylabel(CHANNEL_LABELS[i], fontsize=9)
    tag = 'MODEL INPUT' if MODEL_INPUT[i] else 'VALIDATION ONLY'
    ax.annotate(tag, xy=(0.01, 0.88), xycoords='axes fraction',
                fontsize=8, color='white',
                bbox=dict(boxstyle='round,pad=0.3',
                          fc='#1565C0' if MODEL_INPUT[i] else '#6A1B9A',
                          alpha=0.85))
    ax.grid(True, alpha=0.2)

axes[-1].set_xlabel('Time (ms)', fontsize=10)
plt.tight_layout()
plt.savefig('../outputs/figures/01_zoom_10ms.png', dpi=150, bbox_inches='tight')
plt.show()
print('Saved -> outputs/figures/01_zoom_10ms.png')
````

## 8. Compare all 7 loosening levels — Channel A side by side

Visual proof of why classification is hard at boundary levels (e.g. Class 3 vs 4).

````python
TORQUES   = ['05', '10', '20', '30', '40', '50', '60']
ZOOM_MS   = 5
zoom_samp = int(ZOOM_MS / 1000 * FS)
time_ms   = np.arange(zoom_samp) / FS * 1000

fig, axes = plt.subplots(7, 1, figsize=(16, 16), sharex=True)
fig.suptitle(
    f'Channel A (Micro80) -- All 7 Loosening Levels -- Campaign B -- First {ZOOM_MS}ms',
    fontsize=13, fontweight='bold'
)

for i, torque in enumerate(TORQUES):
    files = sorted(DATA_DIR.rglob(f'*{torque}cNm_B*.mat'))
    if not files:
        axes[i].set_ylabel(f'Class {i+1}\n{torque} cNm\n(NO FILE)', fontsize=8)
        axes[i].text(0.5, 0.5, 'File not found', transform=axes[i].transAxes,
                     ha='center', va='center', color='red')
        continue

    m   = scipy.io.loadmat(files[0])
    sig = m['A'].squeeze().astype(np.float32)

    # Colour gradient: red = loose, green = tight
    r = int(220 * (1 - i / 6))
    g = int(180 * (i / 6))
    color = f'#{r:02X}{g:02X}40'

    axes[i].plot(time_ms, sig[:zoom_samp], color=color, linewidth=0.8)
    axes[i].set_ylabel(f'Class {i+1}\n{torque} cNm', fontsize=9)
    axes[i].grid(True, alpha=0.2)

axes[-1].set_xlabel('Time (ms)', fontsize=10)
plt.tight_layout()
plt.savefig('../outputs/figures/01_all_levels_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
print('Key observation:')
print('  Class 1 (05cNm) -- frequent large bursts -- bolt nearly free')
print('  Class 7 (60cNm) -- mostly quiet -- bolt fully tight')
print('  Adjacent classes look visually similar -- this is why the transformer is needed')
print('Saved -> outputs/figures/01_all_levels_comparison.png')
````

## 9. Windowing preview — what one 2ms model input looks like

````python
WINDOW_SIZE   = 10_000   # 2ms at 5 MHz
WINDOW_STRIDE = 5_000    # 50% overlap

n_windows = (N_SAMPLES - WINDOW_SIZE) // WINDOW_STRIDE + 1
print(f'One file    : {N_SAMPLES:,} samples ({DURATION_S:.3f}s at {FS/1e6:.0f} MHz)')
print(f'Window size : {WINDOW_SIZE:,} samples = {WINDOW_SIZE/FS*1000:.1f} ms')
print(f'Stride      : {WINDOW_STRIDE:,} samples = 50% overlap')
print(f'Windows/file: ~{n_windows:,}')
print(f'Total windows (~340 files): ~{n_windows * 340:,}')

# Plot one window -- 3 AE channels only (model inputs)
window = signals[:3, 0:WINDOW_SIZE]   # shape (3, 10000)
t_ms   = np.arange(WINDOW_SIZE) / FS * 1000

fig, axes = plt.subplots(3, 1, figsize=(14, 7), sharex=True)
fig.suptitle(
    f'One 2ms Window -- 3 AE Model Inputs -- '
    f'Class {info["class_label"]} ({info["torque_str"]} cNm)\n'
    f'Tensor shape fed to CNN Encoders: (3, {WINDOW_SIZE}) -- 3 sensors x {WINDOW_SIZE:,} samples',
    fontsize=11, fontweight='bold'
)

ae_labels = ['Ch A -- Micro80    -> CNN Encoder A',
             'Ch B -- F50A       -> CNN Encoder B',
             'Ch C -- Micro200HF -> CNN Encoder C']

for i, ax in enumerate(axes):
    ax.plot(t_ms, window[i], color=CHANNEL_COLORS[i], linewidth=0.7)
    ax.set_ylabel(ae_labels[i], fontsize=9)
    ax.grid(True, alpha=0.2)

axes[-1].set_xlabel('Time within window (ms)', fontsize=10)
plt.tight_layout()
plt.savefig('../outputs/figures/01_single_window_3sensors.png', dpi=150, bbox_inches='tight')
plt.show()
print(f'\nThis (3, {WINDOW_SIZE:,}) tensor is exactly what gets fed to the 3 CNN encoders.')
print('Saved -> outputs/figures/01_single_window_3sensors.png')
````

## 10. Summary

What we confirmed in this notebook:

- `.mat` files load with `scipy.io.loadmat` — each channel stored as separate key (A, B, C, D)
- Each channel is `(5,111,808, 1)` — slightly more than 1 second at 5 MHz
- Stack into `(4, N)` matrix with `np.stack` + `.squeeze()`
- Filename parsing extracts class label and campaign automatically — no manual labelling
- All 4 channels visualised — 3 AE sensors (model inputs) + vibrometer (validation only)
- Loose bolt (Class 1) produces frequent large voltage bursts — tight bolt (Class 7) mostly quiet
- Adjacent classes look visually similar — single sensor is not enough
- One 2ms window = `(3, 10_000)` tensor — exact model input shape
- ~1,020 windows per file x ~340 files = ~347,000 labelled samples total
